```python
# Import necessary libraries
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg'

print("Libraries imported successfully!")

# Connect to the SQLite database
conn = sqlite3.connect("inventory.db")

# Create a cursor object to interact with the database
cursor = conn.cursor()

# Function to run a SQL query and return the results as a DataFrame
def run_query_to_df(cursor, query):
    cursor.execute(query)
    rows = cursor.fetchall()
    
    # Get column names from cursor
    columns = []
    for col in cursor.description:
        columns.append(col[0])

    # Convert to DataFrame
    df = pd.DataFrame(rows, columns = columns)

    return df
```

```python
query1 = """
SELECT shipping_mode,
    (SELECT COUNT("id") FROM shipping
    WHERE "delivery_status" = "Late delivery" AND shipping_mode = 'First Class')*100.0
    /(SELECT COUNT("id") FROM shipping
    WHERE shipping_mode = 'First Class')*1.0
AS "Late Rate"
FROM shipping
GROUP BY shipping_mode
HAVING shipping_mode = 'First Class'
"""

df1 = run_query_to_df(cursor, query1)

query2 = """
SELECT shipping_mode,
    (SELECT COUNT("id") FROM shipping
    WHERE "delivery_status" = "Late delivery" AND shipping_mode = 'Same Day')*100.0
    /(SELECT COUNT("id") FROM shipping
    WHERE shipping_mode = 'Same Day')*1.0
AS "Late Rate"
FROM shipping
GROUP BY shipping_mode
HAVING shipping_mode = 'Same Day'
"""

df2 = run_query_to_df(cursor, query2)

query3 = """
SELECT shipping_mode,
    (SELECT COUNT("id") FROM shipping
    WHERE "delivery_status" = "Late delivery" AND shipping_mode = 'Second Class')*100.0
    /(SELECT COUNT("id") FROM shipping
    WHERE shipping_mode = 'Second Class')*1.0
AS "Late Rate"
FROM shipping
GROUP BY shipping_mode
HAVING shipping_mode = 'Second Class'
"""

df3 = run_query_to_df(cursor, query3)

query4 = """
SELECT shipping_mode,
    (SELECT COUNT("id") FROM shipping
    WHERE "delivery_status" = "Late delivery" AND shipping_mode = 'Standard Class')*100.0
    /(SELECT COUNT("id") FROM shipping
    WHERE shipping_mode = 'Standard Class')*1.0
AS "Late Rate"
FROM shipping
GROUP BY shipping_mode
HAVING shipping_mode = 'Standard Class'
"""

df4 = run_query_to_df(cursor, query4)

combined_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
combined_df
plt.bar(combined_df["shipping_mode"], combined_df["Late Rate"], color = ['red','blue','blue','green'], width = 0.5)
plt.xlabel("Shipping Mode")
plt.ylabel("Late Delivery Rate (%)")
plt.title("Late Delivery Rate by Shipping Mode")
```
